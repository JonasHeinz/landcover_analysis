```python
import geopandas as gpd

As_ZG = gpd.read_file(r"C:\Users\aebim\Documents\02_Ausbildung\Studium\05_Semester\5230_Geoniformatik_Raumanalyse\Projektarbeit\03_GitHub\data\analysis\arealstatistik\Arealstatistik_Zug.gpkg")

BB_ZG = gpd.read_file(r"C:\Users\aebim\Documents\02_Ausbildung\Studium\05_Semester\5230_Geoniformatik_Raumanalyse\Projektarbeit\03_GitHub\data\analysis\av\BB_Zug.gpkg")
```

```python
As_ZG["x"] = As_ZG.geometry.centroid.x
As_ZG["y"] = As_ZG.geometry.centroid.y
print(As_ZG.head())
```

```python
import sys
sys.path.append("../../")
from helpers.raster_helper import calculate_vectorbox
import pandas as pd

# Convert the GeoDataFrame coordinates to numeric type first

As_ZG['x'] = As_ZG['x'].astype(float)
As_ZG['y'] = As_ZG['y'].astype(float)

Raster_As_Zg = calculate_vectorbox(As_ZG, 100)
# print(As_ZG.dtypes)
# As_ZG.info
```

```python
# ensure CRS is metric (LV95 / EPSG:2056 or similar)
print("As_ZG.crs:", As_ZG.crs)
# reproject if needed, example from WGS84 to LV95:
# As_ZG = As_ZG.to_crs(epsg=2056)

# if As_ZG geometries are polygons use centroids, otherwise geometry if already points
points = As_ZG.geometry.centroid

from helpers.raster_helper import calculate_vectorbox
boxsize = 100  # meters

# create GeoSeries of boxes
boxes = points.apply(lambda p: calculate_vectorbox(p, boxsize))

# Make GeoDataFrame with same attributes + new geometry
As_ZG_Raster = gpd.GeoDataFrame(As_ZG.drop(columns="geometry"), geometry=boxes, crs=As_ZG.crs)
# Or keep everything:
# boxes_gdf = As_ZG.copy()
# boxes_gdf.geometry = boxes

# quick check & save
print(As_ZG_Raster.head())
As_ZG_Raster.to_file(r"C:\Users\aebim\Documents\02_Ausbildung\Studium\05_Semester\5230_Geoniformatik_Raumanalyse\Projektarbeit\03_GitHub\data\analysis\arealstatistik\Arealstatistik_Zug_Boxes1.gpkg", driver="GPKG")

# plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8,8))
As_ZG_Raster.plot(ax=ax, facecolor="none", edgecolor="red")
plt.show()
```

```python
# First ensure same CRS
if As_ZG_Raster.crs != BB_ZG.crs:
    BB_ZG = BB_ZG.to_crs(As_ZG_Raster.crs)

# Get centroids of the raster boxes for point-in-polygon operation
raster_points = As_ZG_Raster.geometry.centroid

# Perform spatial join
joined = gpd.sjoin(
    gpd.GeoDataFrame(geometry=raster_points, crs=As_ZG_Raster.crs),
    BB_ZG[['Art', 'geometry']],
    how='left',
    predicate='within'
)

# Add the joined Art values back to As_ZG_Raster
As_ZG_Raster['Art_AV'] = joined['Art']

# Check results
# print(As_ZG_Raster[['Art_AV']].head())
# print("\nValue counts:")
# print(As_ZG_Raster['Art_AV'].value_counts())
# print("\n")
print(As_ZG_Raster.head())
```

```python
import pandas as pd
ipcc_av_mapping_dict = pd.read_csv(r"C:\Users\aebim\Documents\02_Ausbildung\Studium\05_Semester\5230_Geoniformatik_Raumanalyse\Projektarbeit\03_GitHub\data\analysis\av\mapping_ipcc_av.csv", sep=";" ).set_index("AV_ID")["IPCC_ID"].to_dict()
ipcc_arealstatistik_mapping_dict = pd.read_csv(r"C:\Users\aebim\Documents\02_Ausbildung\Studium\05_Semester\5230_Geoniformatik_Raumanalyse\Projektarbeit\03_GitHub\data\analysis\arealstatistik\mapping_ipcc_arealstatistik.csv", sep=";" ).set_index("AS_ID")["IPCC_ID"].to_dict()
```
