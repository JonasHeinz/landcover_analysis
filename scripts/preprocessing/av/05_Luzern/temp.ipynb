```python
import geopandas as gpd
import os
import matplotlib.pyplot as plt
```

```python
CH_Gesamt = gpd.read_file(r"C:\Users\st1179523\Documents\GitHub\landcover_analysis\data\preprocessing\av\BB_CH_Gesamt.gpkg")

fig, ax = plt.subplots(figsize=(8, 8))
CH_Gesamt.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=0.8)
ax.set_axis_off()
plt.show()
```
